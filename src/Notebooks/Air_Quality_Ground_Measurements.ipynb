```python
def basic_plot(data, x, y, title):
  fig = px.bar(data, x, y, title=title, width=1200, height=500)

  fig.add_vrect(x0='2020-03-25', x1='2020-04-14', annotation_text='L1', annotation_position='top right', fillcolor='green', opacity=0.25, line_width=0)
  fig.add_vrect(x0='2020-04-15', x1='2020-05-03', annotation_text='L2', annotation_position='top right', fillcolor='green', opacity=0.25, line_width=0)
  fig.add_vrect(x0='2020-05-04', x1='2020-05-17', annotation_text='L3', annotation_position='top right', fillcolor='green', opacity=0.25, line_width=0)
  fig.add_vrect(x0='2020-05-18', x1='2020-05-31', annotation_text='L4', annotation_position='top right', fillcolor='green', opacity=0.25, line_width=0)
  fig.add_vrect(x0='2020-06-01', x1='2022-01-31', annotation_text='Unlock', annotation_position='top right', fillcolor='red', opacity=0.25, line_width=0)
  fig.update_layout(
        xaxis=dict(
            rangeselector=dict(
                buttons=list([
                    dict(count=1,
                        label="1m",
                        step="month",
                        stepmode="backward"),
                    dict(count=6,
                        label="6m",
                        step="month",
                        stepmode="backward"),
                    dict(count=1,
                        label="1y",
                        step="year",
                        stepmode="backward"),
                    dict(step="all")
                ])
            ),
            rangeslider=dict(
                visible=True
            ),
            type="date"
        )
    )
  fig.show()
```

```python
data = pd.read_csv('/content/colaba_air_quality.csv', parse_dates=['date'])
data.head()
```

```python
data= data.rename(columns = {" pm25": "pm25", 
                         " pm10":"pm10", 
                         " o3": "o3",
                         ' no2' : 'no2',
                         ' so2' : 'so2',
                         ' co' : 'co'})

data.columns
```

```python
data.info()
```

```python
mask = (data['date'] > '2019-01-27') & (data['date'] <= '2022-01-10')
data = data.loc[mask]
data = data.sort_values(by = 'date')
```

```python
basic_plot(data=data, x='date',y='so2', title='S02 through Time')
```

```python
basic_plot(data=data, x='date',y='no2', title='N02 through Time')
```

```python
basic_plot(data=data, x='date',y='pm25', title='PM2.5 Levels through Time')
```

```python
basic_plot(data=data, x='date',y='co', title='Carbon Monoxide Levels through Time')
```

```python
basic_plot(data=data, x='date',y='o3', title='Ozone Levels through Time')
```

```python
basic_plot(data=data, x='date',y='pm10', title='PM10 Levels through Time')
```
